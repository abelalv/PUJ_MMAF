```python
### C.3.1 Optimización de funciones cuadráticas

import numpy as np
import matplotlib.pyplot as plt

def optimizar_funcion_cuadratica(a, b, c, intervalo):
    """
    Optimiza una función cuadrática de la forma ax^2 + bx + c.
    
    Parámetros:
    a, b, c: Coeficientes de la función cuadrática.
    intervalo: Tupla (inicio, fin) que define el intervalo [a, b].
    
    Retorna:
    Un diccionario con los puntos máximos y mínimos y la gráfica de la función.
    """
    # Definir la función cuadrática
    def f(x):
        return a*x**2 + b*x + c
    
    # Encontrar el vértice de la parábola
    x_vertice = -b / (2*a)
    y_vertice = f(x_vertice)
    
    # Generar valores de x en el intervalo
    x = np.linspace(intervalo[0], intervalo[1], 400)
    y = f(x)
    
    # Graficar la función
    plt.plot(x, y, label=f'$f(x) = {a}x^2 + {b}x + {c}$')
    plt.scatter([x_vertice], [y_vertice], color='red', zorder=5, label='Vértice')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Optimización de una función cuadrática')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Determinar si el vértice es un máximo o un mínimo
    if a > 0:
        tipo = 'mínimo'
    else:
        tipo = 'máximo'
    
    return {
        'vértice': (x_vertice, y_vertice),
        'tipo': tipo
    }

# Ejemplo de uso
a, b, c = -1, 4, 5
intervalo = (-2, 6)
resultado = optimizar_funcion_cuadratica(a, b, c, intervalo)
print("Vértice:", resultado['vértice'])
print("Tipo de punto:", resultado['tipo'])
```

## C.3.2  Graficar funciones polinomicas y polinomicas racionales



```python
import numpy as np

import matplotlib.pyplot as plt

def graficar_polinomio(coeficientes, intervalo):
    """
    Grafica un polinomio de cualquier grado.
    
    Parámetros:
    coeficientes: Lista de coeficientes del polinomio en orden descendente.
    intervalo: Tupla (inicio, fin) que define el intervalo [a, b].
    """
    # Crear el polinomio usando numpy
    p = np.poly1d(coeficientes)
    
    # Generar valores de x en el intervalo
    x = np.linspace(intervalo[0], intervalo[1], 400)
    y = p(x)
    
    # Graficar el polinomio
    plt.plot(x, y, label=f'Polinomio:\n {p}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfica del polinomio')
    plt.legend()
    plt.grid(True)
    plt.show()
    print(p)

# Ejemplo de uso
coeficientes = [1, -3, 2,2,2]  # Coeficientes del polinomio x^2 - 3x + 2
intervalo = (-2, 4)
graficar_polinomio(coeficientes, intervalo)
```

```python
import numpy as np

import matplotlib.pyplot as plt

def graficar_funcion_racional(numerador_coef, denominador_coef, intervalo):
    """
    Grafica una función racional de la forma P(x)/Q(x).
    
    Parámetros:
    numerador_coef: Lista de coeficientes del polinomio del numerador en orden descendente.
    denominador_coef: Lista de coeficientes del polinomio del denominador en orden descendente.
    intervalo: Tupla (inicio, fin) que define el intervalo [a, b].
    """
    # Crear los polinomios usando numpy
    P = np.poly1d(numerador_coef)
    Q = np.poly1d(denominador_coef)
    
    # Generar valores de x en el intervalo
    x = np.linspace(intervalo[0], intervalo[1], 400)
    
    # Evaluar la función racional
    y = P(x) / Q(x)
    
    # Graficar la función racional
    plt.plot(x, y, label=f'R(x) = ({P}) / ({Q})')
    plt.xlabel('x')
    plt.ylabel('R(x)')
    plt.title('Gráfica de la función racional')
    plt.legend()
    plt.grid(True)
    plt.ylim(-10, 10)  # Limitar el eje y para evitar valores extremos
    plt.show()

# Ejemplo de uso
numerador_coef = [1, -3, 2]  # Coeficientes del polinomio del numerador
denominador_coef = [1, -2, 1]  # Coeficientes del polinomio del denominador
intervalo = (-10, 10)
graficar_funcion_racional(numerador_coef, denominador_coef, intervalo)
```

### C.3.3 Gráficar funciones a trozos

```python
import numpy as np

import matplotlib.pyplot as plt

def graficar_funcion_a_trozos(funciones, intervalos):
    """
    Grafica una función a trozos.
    
    Parámetros:
    funciones: Lista de funciones lambda que definen cada trozo.
    intervalos: Lista de tuplas que definen los intervalos de cada trozo.
    """
    x_vals = []
    y_vals = []
    
    for i, (func, intervalo) in enumerate(zip(funciones, intervalos)):
        x = np.linspace(intervalo[0], intervalo[1], 400)
        y = func(x)
        x_vals.extend(x)
        y_vals.extend(y)
        plt.plot(x, y, label=f'Trozo {i+1}')
    
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfica de la función a trozos')
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejemplo de uso
funciones = [
    lambda x: x**2,
    lambda x: -x + 2,
    lambda x: np.sin(x)
]
intervalos = [
    (-10, -2),
    (-2, 2),
    (2, 10)
]

graficar_funcion_a_trozos(funciones, intervalos)
```

### C.3.3 Obtener ceros racionales de funciones polinomicas

```python
import numpy as np
import math

import matplotlib.pyplot as plt

def graficar_polinomio_y_ceros(coeficientes, intervalo):
    """
    Grafica un polinomio de cualquier grado y muestra sus ceros.
    
    Parámetros:
    coeficientes: Lista de coeficientes del polinomio en orden descendente.
    intervalo: Tupla (inicio, fin) que define el intervalo [a, b].
    """
    # Crear el polinomio usando numpy
    p = np.poly1d(coeficientes)
    
    # Generar valores de x en el intervalo
    x = np.linspace(intervalo[0], intervalo[1], 400)
    y = p(x)
    
    # Encontrar los ceros del polinomio
    ceros = np.roots(coeficientes)
    
    # Graficar el polinomio
    plt.plot(x, y, label=f'Polinomio:\n {p}')
    plt.scatter(ceros, np.zeros_like(ceros), color='red', zorder=5, label='Ceros')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfica del polinomio y sus ceros')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Encontrar ceros racionales
    ceros_racionales = [cero for cero in ceros if np.isreal(cero)]
    ceros_racionales = [cero.real for cero in ceros_racionales]
    
    print("Ceros del polinomio:", ceros)
    print("Ceros racionales del polinomio:", ceros_racionales)

# Ejemplo de uso
coeficientes = [1, -3, 2, 2, 2]  # Coeficientes del polinomio
intervalo = (-10, 10)
graficar_polinomio_y_ceros(coeficientes, intervalo)
```
