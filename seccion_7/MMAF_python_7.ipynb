### Funciones trigonometricas básicas

Las funciones trigonometricas son funciones básicas que permiten relacional medidas de ángulos y lados en un triángulo rectángulo. Las funciones trigonometricas básicas son:

* Seno
* Coseno
* Tangente

Estas funciones son periódicas, es decir, se repiten en intervalos regulares. La función seno y coseno tienen un periodo de 2π, mientras que la función tangente tiene un periodo de $\pi$.




### Función trigonométricas desde el circulo trigonometrico

contrario a las manecillas del reloj.

### Implementación de las funciones seno, coseno y tangente en Python

Las funciones trigonométricas `seno`, `coseno` y `tangente` se pueden utilizar en Python mediante los módulos `numpy` o `math`.  
Ambos permiten realizar cálculos trigonométricos, aunque presentan algunas diferencias importantes:

- El módulo `math` trabaja con **valores individuales** (escalares).
- El módulo `numpy` trabaja con **vectores o arreglos** y permite calcular valores de forma masiva, lo cual es útil para graficar.

#### Evaluación de funciones trigonométricas en un punto

Para evaluar el seno, coseno o tangente de un ángulo determinado, puede utilizar cualquiera de los dos módulos.

```python
import numpy as np

angulo = np.pi / 4   # 45 grados en radianes
seno = np.sin(angulo)
print(seno)
```

Con `math`:

```python
import math as mt

angulo = mt.pi / 4
seno = mt.sin(angulo)
print(seno)
```

Ambos producirán el mismo resultado, aunque `numpy` es más adecuado para cálculos sobre vectores o listas de valores.

### Graficación de funciones trigonométricas

Para visualizar el comportamiento de una función trigonométrica, se puede utilizar el módulo `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de valores de x entre -2π y 2π
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Calcular el valor del seno, coseno y tangente para cada punto
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar las tres funciones
plt.plot(x, y1, label='sin(x)', color='b')
plt.plot(x, y2, label='cos(x)', color='r')
plt.plot(x, y3, label='tan(x)', color='g')

# Ajustar el rango del eje y para que la tangente no distorsione la vista
plt.ylim(-5, 5)

# Título y etiquetas
plt.title("Funciones Seno, Coseno y Tangente")
plt.xlabel("x (radianes)")
plt.ylabel("Valor de la función")
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
```

**Explicación:**

1. `np.linspace(-2*np.pi, 2*np.pi, 1000)` genera 1000 puntos igualmente espaciados entre $-2\pi$ y $2\pi$, cubriendo varios ciclos de las funciones trigonométricas 

2. `np.sin(x)`, `np.cos(x)` y `np.tan(x)` calculan respectivamente los valores del **seno**, **coseno** y **tangente** para cada punto del arreglo `x`. 

3. `plt.figure(figsize=(10, 6))` define el tamaño de la figura donde se dibujarán las tres funciones.
4. `plt.plot(x, y1, ...)`, `plt.plot(x, y2, ...)` y `plt.plot(x, y3, ...)` trazan las tres curvas en diferentes colores y con etiquetas para distinguirlas.
5. `plt.ylim(-5, 5)` limita el rango del eje **y** para que los valores grandes de la tangente no distorsionen la visualización.
6. `plt.title()`, `plt.xlabel()`, `plt.ylabel()` y `plt.legend()` añaden información descriptiva y una leyenda para identificar las funciones.
7. `plt.grid(True)` agrega una cuadrícula que facilita la lectura de la gráfica.
8. `plt.show()` finalmente muestra la figura en pantalla.

### Identidades trigonométricas

Las **identidades trigonométricas** son relaciones entre las funciones seno, coseno, tangente y sus recíprocas, que **se cumplen para cualquier valor del ángulo** (siempre que las expresiones estén definidas).  
Estas identidades son fundamentales en matemáticas y física, pues permiten **simplificar expresiones**, **resolver ecuaciones** y **comprender las propiedades geométricas del círculo unitario**.

#### 1. Identidad pitagórica

A partir del **círculo unitario**, sabemos que todo punto sobre su circunferencia cumple la ecuación:

$$x^2 + y^2 = 1$$

Si identificamos:
\[
x = \cos(\theta) \quad \text{y} \quad y = \sin(\theta)
\]

entonces obtenemos la **identidad pitagórica**:

$$
\sin^2(\theta) + \cos^2(\theta) = 1
$$

Esta relación muestra que los valores de $\sin(\theta)$ y $\cos(\theta)$ siempre están conectados: ninguno puede cambiar sin afectar al otro.

##### Ejercicio

A partir de la identidad pitagórica, divide ambos lados por $\sin^2(\theta)$ y luego por $\cos^2(\theta)$ para obtener dos nuevas identidades:

$$
\tan^2(\theta) + 1 = \sec^2(\theta)
$$

$$
1 + \cot^2(\theta) = \csc^2(\theta)
$$

Estas fórmulas son especialmente útiles para resolver ecuaciones trigonométricas y analizar relaciones entre las funciones recíprocas.

#### 2. Identidades de ángulos complementarios

Dos ángulos son **complementarios** si su suma es de \( 90^\circ \) o \( \frac{\pi}{2} \) radianes.  
Las funciones seno y coseno de ángulos complementarios están relacionadas de la siguiente forma:

$$
\sin(90^\circ - \theta) = \cos(\theta)
\quad \text{y} \quad
\cos(90^\circ - \theta) = \sin(\theta)
$$

Estas igualdades reflejan una simetría fundamental del círculo unitario:  
el seno de un ángulo es igual al coseno de su complemento.

#### 3. Identidades de suma y diferencia de ángulos

Cuando se trabaja con combinaciones de ángulos, es útil expresar las funciones trigonométricas de una suma o diferencia en términos de ángulos más simples:

$$
\sin(A \pm B) = \sin(A)\cos(B) \pm \cos(A)\sin(B)
$$

$$
\cos(A \pm B) = \cos(A)\cos(B) \mp \sin(A)\sin(B)
$$

Estas identidades son esenciales en muchas aplicaciones: desde el análisis de ondas hasta el procesamiento de señales o la resolución de triángulos.

Con la librería 'sympy' podemos comprobar que algunas de las identidades trigonométricas son verdaderas. A continuación se muestra un ejemplo de como se puede comprobar la identidad trigonométrica,

$$\sin^2(\theta) + \cos^2(\theta) = 1,$$



Para ello usamos la función `simplify` de la librería `sympy` que simplifica la expresión matemática.

```python
import sympy as sp

x= sp.symbols('x')
identidad_1 = sp.sin(x)**2 + sp.cos(x)**2
identidad_2 = 1
sp.simplify(identidad_1 - identidad_2)
```

si arroja 0, la identidad es verdadera, en otro contrario es falsa.

Note que en este caso debemos usar las funciones `sin` y `cos` de la librería `sympy`, es decir para llamarlas debemos usar `sp.sin` y `sp.cos`.


### Funciones inversas


Recuerde que para construir una función inversas se necesita la función sea inyectiva y sobreyectiva, como podemos ver en el caso de la función $\sin\theta$ no cumple estas propiedades, por tanto no tiene una función inversa, pero para construir una función inversa debemos restringir el dominio de la función $\sin\theta$, para que cumpla las propiedades necesarias para tener una función inversa. Por tanto,en el caso de la función $\sin\theta$ se restringe entre el intervalo 

$$\Big[-\frac{\pi}{2},\frac{\pi}{2}\Big],$$

y es denotada como $\sin^{-1}\theta$ o $\arcsin\theta$.

En el caso del coseno se restringe entre el intervalo 

$$ \Big[0,\pi\Big],$$

y es denotada como $\cos^{-1}\theta$ o $\arccos\theta$.

En el caso de la tangente se restringe entre el intervalo

$$ \Big[-\frac{\pi}{2},\frac{\pi}{2}\Big],$$

y es denotada como $\tan^{-1}\theta$ o $\arctan\theta$.

### Resolución Gráfica de Ecuaciones Trigonométricas

Las **ecuaciones trigonométricas** son aquellas que involucran funciones trigonométricas de una variable, normalmente un ángulo.  
Resolver una ecuación trigonométrica significa **encontrar los valores del ángulo** que satisfacen la igualdad dada.  
Una forma intuitiva de hacerlo es mediante su **interpretación gráfica**, observando los puntos donde la función trigonométrica corta una línea horizontal.

Por ejemplo, consideremos la ecuación:

$$
\sin(\theta) = A
$$

Sabemos que el seno está definido en el rango \([-1, 1]\), por lo tanto, solo existen soluciones reales si:

$$
-1 \leq A \leq 1
$$

La gráfica de esta ecuación corresponde a los puntos de intersección entre la función seno y la recta \(y = A\).

### Ejemplo: Resolución gráfica de una ecuación trigonométrica

Consideremos la siguiente ecuación trigonométrica cuadrática en $\sin(x)$:

$$
2\sin^2(x) - \sin(x) - 1 = 0
$$

Para resolverla, podemos seguir un **enfoque algebraico** o un **enfoque gráfico**.

#### Enfoque algebraico

Si sustituimos $u = \sin(x)$, la ecuación se transforma en:

$$
2u^2 - u - 1 = 0
$$

Resolviendo la ecuación cuadrática para $u$:

$$
u = 1 \quad \text{o} \quad u = -\frac{1}{2}
$$

Luego, al volver a la variable original $x$:

$$
\sin(x) = 1 \quad \Rightarrow \quad x = \frac{\pi}{2}, \, ... 
$$

$$
\sin(x) = -\frac{1}{2} \quad \Rightarrow \quad x = \frac{7\pi}{6}, \; \frac{11\pi}{6}, \, ...
$$

#### Enfoque gráfico

Podemos verificar estas soluciones representando gráficamente la función:

$$
f(x) = 2\sin^2(x) - \sin(x) - 1
$$

y observando dónde corta al eje $x$ (es decir, donde $f(x) = 0$).



```python
from MMAF import graficar_funcion_trigonometrica

graficar_funcion_trigonometrica()
```

#### Interpretación

En la gráfica se observa que las intersecciones con el eje $x$ ocurren precisamente en los puntos:

$$
x = \frac{\pi}{2}, \quad x = \frac{7\pi}{6}, \quad x = \frac{11\pi}{6}
$$

Estos valores corresponden a las soluciones de la ecuación trigonométrica inicial en el intervalo $[0, 2\pi]$.

Este enfoque permite visualizar de forma clara **cómo las raíces de la ecuación se relacionan con los valores del seno** y **cómo las identidades trigonométricas pueden transformarse en expresiones algebraicas equivalentes**.
