### Funciones trigonometricas básicas

Las funciones trigonometricas son funciones básicas que permiten relacional medidas de ángulos y lados en un triángulo rectángulo. Las funciones trigonometricas básicas son:

* Seno

* Tangente

### Identidades trigonometricas


Las identidades trigonométricas son ecuaciones que son verdaderas para cualquier valor del ángulo, siempre que los lados de la ecuación estén definidos. Estas identidades juegan un papel crucial en la simplificación de expresiones trigonométricas y en la resolución de ecuaciones trigonométricas.

## Identidades Básicas
Una de las identidades fundamentales que deriva directamente del círculo unitario es la **identidad pitagórica**, la cual establece que para cualquier ángulo $\theta$:


$$\sin^2(\theta) + \cos^2(\theta) = 1$$

Esta identidad nos dice que, al sumar los cuadrados de las funciones seno y coseno de un ángulo, siempre obtenemos 1.

### Identidades de Ángulos Complementarios
Otra identidad importante está relacionada con los ángulos complementarios. Dos ángulos son complementarios si suman $90^\circ$ (o $\frac{\pi}{2}$ radianes). Para un ángulo $\theta$, se tiene que:


$$\sin(90^\circ - \theta) = \cos(\theta) \quad \text{y} \quad \cos(90^\circ - \theta) = \sin(\theta)$$

Estas identidades muestran la relación inversa entre las funciones seno y coseno para ángulos complementarios.

### Identidades de Suma y Diferencia de Ángulos
Las identidades de suma y diferencia de ángulos son útiles cuando se trabaja con combinaciones de ángulos. Por ejemplo, para dos ángulos $A$ y $B$:


$$\sin(A \pm B) = \sin(A) \cos(B) \pm \cos(A) \sin(B)$$
$$\cos(A \pm B) = \cos(A) \cos(B) \mp \sin(A) \sin(B)$$

Estas fórmulas nos permiten expresar el seno y coseno de la suma o la diferencia de dos ángulos en términos de las funciones seno y coseno de los ángulos individuales.

### Funciones inversas


Recuerde que para construir una función inversas se necesita la función sea inyectiva y sobreyectiva, como podemos ver en el caso de la función $\sin\theta$ no cumple estas propiedades, por tanto no tiene una función inversa, pero para construir una función inversa debemos restringir el dominio de la función $\sin\theta$, para que cumpla las propiedades necesarias para tener una función inversa. Por tanto,en el caso de la función $\sin\theta$ se restringe entre el intervalo 

$$\Big[-\frac{\pi}{2},\frac{\pi}{2}\Big],$$

y es denotada como $\sin^{-1}\theta$ o $\arcsin\theta$.

En el caso del coseno se restringe entre el intervalo 

$$ \Big[0,\pi\Big],$$

y es denotada como $\cos^{-1}\theta$ o $\arccos\theta$.

En el caso de la tangente se restringe entre el intervalo

$$ \Big[-\frac{\pi}{2},\frac{\pi}{2}\Big],$$

y es denotada como $\tan^{-1}\theta$ o $\arctan\theta$.

## Contexto

En esta experiencia, vamos a modelar la temperatura promedio anual en un país con estaciones usando ondas sinusoidales. Las estaciones del año, como invierno, primavera, verano y otoño, afectan las temperaturas, las cuales pueden variar de manera predecible. Usaremos una función sinusoidal para representar esta variación de temperatura a lo largo del año.

La fórmula que utilizaremos es:

$$ T(t) = A \cdot \sin\left( \frac{2 \pi}{P} \cdot (t - \phi) \right) + C $$

Donde:
- \( T(t) \) es la temperatura en el mes \( t \) (en grados Celsius).
- \( A \) es la amplitud (la diferencia entre la temperatura más alta y la más baja dividida por 2).
- \( P \) es el periodo (en este caso, 12 meses).
- \( t \) es el mes del año.
- \( \phi \) es la fase (para ajustar el desplazamiento de las estaciones).
- \( C \) es la temperatura media anual.


Vamos a usar el siguiente código para modelar las variaciones de temperatura durante el año en un país con estaciones. Este código genera una gráfica que muestra cómo cambia la temperatura a lo largo de los 12 meses.


````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
A = 10  # Amplitud: diferencia entre la temperatura media de verano e invierno dividida entre 2
P = 12  # Periodo: 12 meses en un año
phi = 2  # Fase: para ajustar el ciclo al inicio del año
C = 15  # Temperatura media anual en grados Celsius

# Generar el eje del tiempo (en meses)
t = np.linspace(0, 12, 1000)  # 12 meses, con 1000 puntos para suavizar la curva

# Función de temperatura basada en un modelo sinusoidal
T = A * np.sin(2 * np.pi / P * (t - phi)) + C

# Generar la gráfica
plt.figure(figsize=(10, 6))
plt.plot(t, T, label='Temperatura modelada')
plt.axhline(C, color='red', linestyle='--', label='Temperatura media anual')
plt.xlabel('Meses')
plt.ylabel('Temperatura (°C)')
plt.title('Modelo de temperatura promedio anual en un país con estaciones')
plt.grid(True)

# Etiquetar meses del año
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
plt.xticks(np.arange(0, 12, step=1), meses, rotation=45)  # Ajustamos a 12 valores

plt.legend()
plt.show()
````

Las funciones seno y coseno son la base de muchas aplicaciones en la ingeniería, como la generación de sonido, la transmisión de señales, y el análisis de vibraciones. En este ejercicio, aprenderás cómo se construyen estas ondas y cómo se combinan para formar una melodía.

2. **¿Qué es la frecuencia de una onda?** Cada nota en la canción tiene una frecuencia diferente. La frecuencia es el número de veces que una onda completa un ciclo en un segundo, y se mide en Hertz (Hz). Por ejemplo, la nota Mi (E) tiene una frecuencia de 329.63 Hz, lo que significa que su onda completa 329.63 ciclos en un segundo. Si una onda tiene una frecuencia de 2 Hz, ¿cuántos ciclos completa en un segundo? ¿Cuántos ciclos completaría en 2 segundos?

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from IPython.display import Audio

# Parámetros del sonido
rate = 22050  # Tasa de muestreo en Hz
duration = 0.5  # Duración de cada nota en segundos
t = np.linspace(0, duration, int(rate * duration), endpoint=False)  # Eje de tiempo

# Frecuencias de las notas musicales (en Hz)
notes = {
    'C': 261.63,  # Do
    'D': 293.66,  # Re
    'E': 329.63,  # Mi
    'F': 349.23,  # Fa
    'G': 392.00,  # Sol
    'A': 440.00,  # La
    'B': 493.88,  # Si
    'C_high': 523.25  # Do (una octava más alta)
}

# Función para generar una onda sinusoidal para una nota
def generate_wave(note, duration=0.5, rate=22050):
    t = np.linspace(0, duration, int(rate * duration), endpoint=False)
    frequency = notes[note]
    wave = 0.5 * np.sin(2 * np.pi * frequency * t)  # Amplitud de 0.5 para evitar saturación
    return wave

# Melodía de "Los Pollitos" (simplificada)
melody = [
    'E', 'E', 'F', 'G', 'G', 'F', 'E', 'D',  # "Los pollitos dicen pío, pío, pío"
    'C', 'C', 'D', 'E', 'E', 'D', 'D',       # "cuando tienen hambre"
    'E', 'E', 'F', 'G', 'G', 'F', 'E', 'D',  # "y cuando tienen frío"
    'C', 'C', 'D', 'E', 'D', 'C'             # "La mamá les busca el maíz y el trigo"
]

# Generar la onda completa de la canción concatenando las notas
song = np.concatenate([generate_wave(note) for note in melody])

# Asegurarse de que el sonido dure 10 segundos (recortando o rellenando con ceros si es necesario)
desired_duration = 10  # en segundos
song_length = rate * desired_duration

if len(song) < song_length:
    # Si la canción es más corta que 10 segundos, rellenar con ceros (silencio)
    song = np.pad(song, (0, song_length - len(song)), 'constant')
else:
    # Si es más larga, recortar
    song = song[:song_length]

# Escuchar la canción
Audio(song, rate=rate)

# (Opcional) Guardar el archivo de audio en formato .wav
wav.write('los_pollitos.wav', rate, (song * 32767).astype(np.int16))
````
